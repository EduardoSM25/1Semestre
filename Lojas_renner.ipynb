```python
import pandas as pd
import folium
from folium.plugins import HeatMap
from branca.colormap import LinearColormap
```

```python
lojas_renner = pd.read_excel('/content/lojas_renner.xlsx')
lojas_renner
```

```python
coordenadas_lojas = lojas_renner[['latitude', 'longitude']].values
coordenadas_lojas
```

```python
mapa = folium.Map(location = [-15.77972, -47.92972], zoom_start = 6)
folium.TileLayer(tiles = 'cartodbpositron').add_to(mapa)
```

```python
qtd_lojas= len(lojas_renner)

colormap = LinearColormap(
    colors=['blue', 'green', 'yellow', 'orange', 'red'],
    vmin=0, vmax=qtd_lojas
)
colormap.caption = 'Quantidade de lojas'
colormap.add_to(mapa)
```

```python
HeatMap(coordenadas_lojas, radius=20, blur = 10, show_scale = True).add_to(mapa)

mapa.save('heatmap_renner.html')
```
